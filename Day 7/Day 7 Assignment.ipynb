## Performing statistical testings

```python
dataset = pd.read_csv("general_data.csv")
dataset.head()
print(dataset['Attrition'].dtype)
```

### attrition: with respect to distance from home

```python
from scipy.stats import pearsonr
```

```python
print(dataset['Attrition'].dtype)
dataset['Attrition'] = dataset['Attrition'].apply(lambda x: 1.0 if x == 'Yes' else 0.0)
dataset['Attrition'].head()
dataset.head()
```

```python
a = dataset.DistanceFromHome
b = dataset.Attrition
stats, p = pearsonr(b,a)
```

```python
print(stats, p)
```
